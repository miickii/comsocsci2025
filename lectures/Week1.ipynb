### 3.1
Inspect the HTML of the page and use web-scraping to get the names of all researchers that contributed to the conference in 2023. The goal is the following: (i) get as many names as possible including: keynote speakers, chairs, authors of parallel talks and authors of posters; (ii) ensure that the collected names are complete and accurate as reported in the website (e.g. both first name and family name); (iii) ensure that no name is repeated multiple times with slightly different spelling.

```python
import requests
from bs4 import BeautifulSoup
import re

# URL of the conference page
url = "https://ic2s2-2023.org/program"

# Fetch the page
response = requests.get(url)
html_content = response.text

# Parse with BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")
```

```python
# Find all sections with id="main"
sections = soup.find_all("section", id="main")

# Initialize a set to store unique names
researcher_names_2023 = set()

# Regular expression to match names inside <i> and <u>
name_pattern = re.compile(r"\b[A-Z][a-z]+(?:\s[A-Z]\.?)?(?:\s[A-Z][a-z]+)+\b")

for section in sections:
    # Find all <i> tags (names appear here)
    for i_tag in section.find_all("i"):
        # Extract text
        text = i_tag.get_text()
        
        # Extract potential names
        matches = name_pattern.findall(text)
        for match in matches:
            researcher_names_2023.add(match)

# Print all unique researcher names
print(len(researcher_names_2023))
print("\n".join(sorted(researcher_names_2023)))
```

```python
import pandas as pd
import re

files = ["files/IC2S2_2024_lightning_talks.csv", "files/IC2S2_2024_oral_panels.csv", "files/IC2S2_2024_posters.csv"]
file_path = "files.csv"  # Change to .xlsx if needed

authors_df = pd.DataFrame(columns=["author"])

# Iterate through each file
for file in files:
    # Read CSV
    df = pd.read_csv(file)

    # Find the column that contains "author" (case-insensitive)
    author_col = [col for col in df.columns if "author" in col.lower()]
    
    if author_col:  # If a matching column is found
        author_col = author_col[0]  # Extract the column name
        temp_df = df[[author_col]].rename(columns={author_col: "author"})  # Rename to "author"
        authors_df = pd.concat([authors_df, temp_df], ignore_index=True) 

# Extract authors
researcher_names_2024 = set()  # Use a set to store unique names

for entry in authors_df["author"].dropna():  # Drop empty cells
    # Split names (assuming names are separated by commas or semicolons)
    names = entry.split(";")
    
    # Clean names (remove extra spaces, institutes, and keep only names)
    for name in names:
        name = name.strip()  # Remove extra spaces
        name = name.split(",")[0]  # Remove institutes
        if name:
            researcher_names_2024.add(name)

print(len(researcher_names_2024))
print("\n".join(sorted(list(researcher_names_2024))))
```

```python
returning = researcher_names_2023 & researcher_names_2024

# Researchers who are new in 2024
new_in_2024 = researcher_names_2024 - researcher_names_2023

# Researchers who attended in 2023 but not in 2024
not_in_2024 = researcher_names_2023 - researcher_names_2024

# Summary statistics
summary = {
    "Returning researchers": len(returning),
    "New researchers in 2024": len(new_in_2024),
    "Researchers not attending in 2024": len(not_in_2024),
}

print(summary)
```
